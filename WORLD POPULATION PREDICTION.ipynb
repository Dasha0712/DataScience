```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r'pop.csv')
data
```

```python
data.head()
```

```python
alldata = ['object','float','int']
```

```python
data = data.fillna(method='ffill')
```

```python
print(data.describe(include = alldata))
```

```python
y=data["pop2020"]
x=data[["yearlychange","netchange","density","landarea","migrants","fertrate","medage","urbanpop","worldshare"]]
```

```python
x
```

```python
y
```

```python
x1 = sm.add_constant(x)
```

```python
results = sm.OLS(y,x1).fit()
```

```python
np.asarray(data)
```

```python
results.summary()
```

```python
parameter = results.params
```

```python
data1 = list(data)
```
